```python
from qiskit import *
```

```python
%matplotlib inline
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
secretNumber = '101110101'
l = len(secretNumber)
```

```python
qcirc = QuantumCircuit(l+1, l)
```

```python
qcirc.x(l)
```

```python
qcirc.h(range(l+1))
```

```python
qcirc.barrier()
qcirc.draw(output='mpl')
```

```python
for index in range(l):
    if secretNumber[l-index-1] == '1':
        qcirc.cx(index, l)
```

```python
qcirc.barrier()
qcirc.draw(output='mpl')
```

```python
qcirc.h(range(l+1))
```

```python
qcirc.barrier()
qcirc.draw(output='mpl')
```

```python
qcirc.measure(range(l), range(l))
```

```python
qcirc.barrier()
qcirc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(qcirc, backend=simulator, shots=1).result()
```

```python
plot_histogram(result.get_counts(qcirc))
```
