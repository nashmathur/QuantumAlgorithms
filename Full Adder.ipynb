```python
from qiskit import *
```

```python
%matplotlib inline
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
qcirc = QuantumCircuit(6,2)
```

```python
qcirc.x([1])
```

```python
qcirc.barrier()
qcirc.draw(output='mpl')
```

```python
qcirc.cx(0,3)
qcirc.cx(1,3)
qcirc.cx(2,3)
```

```python
qcirc.ccx(0,1,4)
qcirc.cx(2,5)
qcirc.cx(4,5)
qcirc.ccx(4,2,5)
```

```python
qcirc.measure([3,5],[0,1])
```

```python
qcirc.barrier()
qcirc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(qcirc, backend=simulator).result()
```

```python
plot_histogram(result.get_counts(qcirc))
```
